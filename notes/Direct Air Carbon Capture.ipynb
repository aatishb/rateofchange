# Doing the Math on Direct Air Carbon Capture

Let's calculate the energy efficiency of an ideal carbon capture device.

The initial state of the air has CO₂ molecules (red) and other molecules (grey) mixed.



The final state of the air after running the machine has CO₂ molecules separated into a compartment.



The change in entropy is provided by

$\Delta S = n_1 R ln \frac{V_1}{V} + n_2 R ln \frac{V_2}{V}$

where $n_1$ and $n_2$ are the number of moles of CO₂ and non-CO₂ molecules, $R$ is the gas constant, $V_1/V$ and $V_2/V$ are the fraction of the total volume taken up by CO₂ and non-CO₂ molecules when separated.

If both compartments are at the same temperature and pressure as the original air, then we can use the gas law to show that
$\frac{n_1}{V_1} = \frac{n_2}{V_2} = \frac{n}{V}$
i.e.
$\frac{V_1}{V} = \frac{n_1}{n}$ and $\frac{V_2}{V} = \frac{n_2}{n}$

i.e. the volume of each compartment is in proportion to its relative concentration. Using this fact, the change in entropy simplifies to

$\Delta S = n R \text{ }(\text{ }x \text{ }ln x + (1-x)\text{ } ln (1-x)\text{ })$

where $x$ refers to the fraction of CO₂ in the air. This is the the [entropy of mixing](https://en.wikipedia.org/wiki/Entropy_of_mixing) formula.

To extract 1 mole of CO₂, we need to extract 1/x moles of air.

So the change in entropy per mole of CO₂ is

$\Delta S = R\text{ } (\text{ }x \text{ }ln x + (1-x)\text{ } ln (1-x)\text{ }) \text{ }/\text{ } x$

The least amount of energy needed to bring about this change [is](https://en.wikipedia.org/wiki/Entropy_of_mixing#Gibbs_free_energy_of_mixing) 

$\Delta E = -T \Delta S$

So the least amount of energy needed to extract 1 mole of CO₂ is

$\Delta E = - R\text{ }T \text{ } (\text{ }x \text{ }ln x + (1-x)\text{ } ln (1-x)\text{ }) \text{ }/\text{ } x$

Plugging in $R = 8.3 J/K$ and $T = 300K$ and $x = 400 \text{ ppm} = \frac{400}{10^6} = 0.0004$, [we find](https://www.wolframalpha.com/input/?i=-+8.3+*+300+*+%28%281-x%29*ln%281-x%29+%2B+x*ln%28x%29%29%2Fx%2C+x+%3D+0.0004) the minimum energy of extracting CO₂ is ~ 22 kJ/mole, which is within 10% of [published](https://pubs.acs.org/doi/pdf/10.1021/acs.chemrev.6b00173) [estimates](https://pubs.rsc.org/en/content/articlelanding/2014/ee/c3ee42350f) of ~20 kJ/mole.

For flue gas at a natural gas or coal power station, the CO₂ concentration is 5% and 15%, respectively, and the air temperature is 65C i.e. 338K. Plugging in these value, [we](https://www.wolframalpha.com/input/?i=-+8.3+*+338+*+%28%281-x%29*ln%281-x%29+%2B+x*ln%28x%29%29%2Fx%2C+x+%3D+0.05) [find](https://www.wolframalpha.com/input/?i=-+8.3+*+338+*+%28%281-x%29*ln%281-x%29+%2B+x*ln%28x%29%29%2Fx%2C+x+%3D+0.15) the minimum energy needed to extract CO₂ from flue gas to be ~11 kJ/mole for natural gas and ~8kJ/mole for coal. (Compared to [published estimates](https://www.researchgate.net/profile/Niall_Mac_Dowell/publication/265122182_Carbon_capture_and_storage_update/links/53ff20610cf283c3583c8007/Carbon-capture-and-storage-update.pdf) of 8.4 kJ/mole for natural gas and 5.3 kJ/mole for coal).

This suggests a roughly 2-2.5 increase in carbon capture efficiency by extracting CO₂ at a fossil fuel power plant source compared to direct air extraction.

For more careful, detailed calculations of the thermodynamic limits of air capture, see [this paper](https://www.sciencedirect.com/science/article/abs/pii/S0360544212006901).

```python
# importing the modules 
import numpy as np  
import matplotlib.pyplot as plt  

# data to be plotted 
x = np.linspace(0,1,10000)
R = 8.3
T = 300
conversion = 1 / 44 # 1 mole of CO2 = 44 g
y = - conversion * R * T * (x * np.log(x) + (1-x)*np.log(1-x))/x
  
# plotting 
my_dpi = 120
plt.figure(figsize=(600/my_dpi, 400/my_dpi), dpi=my_dpi)
plt.title("Minimum Energy Cost to Extract a Tonne of CO₂")  
plt.xlabel("Fraction of CO₂ in the Air")  
plt.ylabel("Energy (in MegaJoules)")  
plt.plot(x, y, color ="red") 
plt.show()
```
